```python
import json
from pathlib import Path

def extract_improved_predictions(dpo_file, origin_file, output_dir="comparison_10"):
    dpo_data = []
    with open(dpo_file, 'r', encoding='utf-8') as f:
        for line in f:
            dpo_data.append(json.loads(line))
    
    origin_data = []
    with open(origin_file, 'r', encoding='utf-8') as f:
        for line in f:
            origin_data.append(json.loads(line))
    

    dpo_improved = []
    origin_missed = []
    
    for i, (dpo_item, origin_item) in enumerate(zip(dpo_data, origin_data)):
        dpo_success = dpo_item.get("success_10", dpo_item["recall"][4]) == 1
        origin_success = origin_item.get("success_10", origin_item["recall"][4]) == 1
        
        if dpo_success and not origin_success:
            dpo_improved.append(dpo_item)
            origin_missed.append(origin_item)
    

    output_path = Path(output_dir)
    output_path.mkdir(exist_ok=True, parents=True)
    

    dpo_improved_file = output_path / "dpo_improved.jsonl"
    with open(dpo_improved_file, 'w', encoding='utf-8') as f:
        for item in dpo_improved:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")
    
    origin_missed_file = output_path / "origin_missed.jsonl"
    with open(origin_missed_file, 'w', encoding='utf-8') as f:
        for item in origin_missed:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")
    
    print(f"总样本数: {len(dpo_data)}")
    print(f"DPO模型改进的样本数 (DPO命中而原始模型未命中): {len(dpo_improved)}")
    print(f"改进的样本已保存至:")
    print(f"- DPO结果: {dpo_improved_file}")
    print(f"- 原始结果: {origin_missed_file}")
    
    return len(dpo_improved)

# 使用示例
if __name__ == "__main__":
    dpo_file = "test_predictions_dpo_final.jsonl"
    origin_file = "test_predictions_7b_origin.jsonl"
    extract_improved_predictions(dpo_file, origin_file)
```
